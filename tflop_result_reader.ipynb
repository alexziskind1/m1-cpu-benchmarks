```python
import json
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
with open("./results.json") as f:
    results = json.load(f)
```

```python
x_axis = [8, 32, 128, 1024, 2048, 4096, 8192]
m1_conda_fp32 = results["m1_conda"]["single"]
m1_accelerate_fp32 = results["m1_accelerate"]["single"]
m1max_accelerate_fp32 = results["m1max_accelerate"]["single"]
m1max_conda_fp32 = results["m1max_conda"]["single"]
m1_accelerate_fp64 = results["m1_accelerate"]["double"]
m1_conda_fp64 = results["m1_conda"]["double"]
m1max_accelerate_fp64 = results["m1max_accelerate"]["double"]
m1max_conda_fp64 = results["m1max_conda"]["double"]
m2max_accelerate_fp32 = results["m2max_accelerate"]["single"]
m2max_accelerate_fp64 = results["m2max_accelerate"]["double"]
m3max_accelerate_fp32 = results["m3max_accelerate"]["single"]
m3max_accelerate_fp64 = results["m3max_accelerate"]["double"]
ryzen_fp32 = results["5600X"]["single"]
ryzen_fp64 = results["5600X"]["double"]
```

```python
plt.plot(x_axis, m1_accelerate_fp32, label="M1 accelerate") 
plt.plot(x_axis, m1_conda_fp32, label="M1 conda")
plt.plot(x_axis, m1max_accelerate_fp32, label="M1 Max accelerate") 
plt.plot(x_axis, m1max_conda_fp32, label="M1 Max conda")
plt.plot(x_axis, m2max_accelerate_fp32, label="M2 Max accelerate") 
plt.plot(x_axis, m3max_accelerate_fp32, label="M3 Max accelerate") 
plt.plot(x_axis, ryzen_fp32, label="5600X")
plt.legend()
plt.ylim([0,2.8])
plt.title("NumPy MatMul FP32")
plt.ylabel("TFLOPS")
plt.xlabel("Matrix Size")
plt.show()
```

```python
plt.plot(x_axis, m1_accelerate_fp64, label="M1 accelerate") 
plt.plot(x_axis, m1_conda_fp64, label="M1 conda")
plt.plot(x_axis, m1max_accelerate_fp64, label="M1 Max accelerate") 
plt.plot(x_axis, m1max_conda_fp64, label="M1 Max conda")
plt.plot(x_axis, m2max_accelerate_fp64, label="M2 Max accelerate")
plt.plot(x_axis, m3max_accelerate_fp64, label="M3 Max accelerate")
plt.plot(x_axis, ryzen_fp64, label="5600X")
plt.legend()
plt.ylim([0,0.8])
plt.title("NumPy MatMul FP64")
plt.ylabel("TFLOPS")
plt.xlabel("Matrix Size")
plt.show()
```

```python
m1_accelerate_fp16 = results["m1_accelerate"]["half"]
m1_conda_fp16 = results["m1_conda"]["half"]
m1max_accelerate_fp16 = results["m1max_accelerate"]["half"]
m1max_conda_fp16 = results["m1max_conda"]["half"]
m2max_accelerate_fp16 = results["m2max_accelerate"]["half"]
m3max_accelerate_fp16 = results["m3max_accelerate"]["half"]
ryzen_fp16 = results["5600X"]["half"]
x_axis_fp16 = x_axis[:len(m1max_conda_fp16)]

plt.plot(x_axis_fp16, m1_accelerate_fp16, label="M1 accelerate") 
plt.plot(x_axis_fp16, m1_conda_fp16, label="M1 conda") 
plt.plot(x_axis_fp16, m1max_accelerate_fp16, label="M1 Max accelerate") 
plt.plot(x_axis_fp16, m1max_conda_fp16, label="M1 Max conda") 
plt.plot(x_axis_fp16, m2max_accelerate_fp16, label="M2 Max accelerate") 
plt.plot(x_axis_fp16, m3max_accelerate_fp16, label="M3 Max accelerate") 
plt.plot(x_axis_fp16, ryzen_fp16, label="5600X") 
plt.legend()
plt.ylim([0,0.0014])
plt.title("NumPy MatMul FP16")
plt.ylabel("TFLOPS")
plt.xlabel("Matrix Size")
plt.show()
```
