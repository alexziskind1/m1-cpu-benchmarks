```python
import json
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
with open("./results.json") as f:
    results = json.load(f)
```

```python
x_axis = [8, 32, 128, 1024, 2048, 4096, 8192]
accelerate_fp32 = results["accelerate"]["single"]
conda_fp32 = results["conda"]["single"]
accelerate_fp64 = results["accelerate"]["double"]
conda_fp64 = results["conda"]["double"]
```

```python
plt.plot(x_axis, accelerate_fp32, label="accelerate fp32") 
plt.plot(x_axis, conda_fp32, label="conda fp32") 
plt.legend()
plt.ylim([0,2.4])
plt.title("NumPy MatMul FP32")
plt.ylabel("TFLOPS")
plt.xlabel("Matrix Size")
plt.show()
```

```python
plt.plot(x_axis, accelerate_fp64, label="accelerate fp64") 
plt.plot(x_axis, conda_fp64, label="conda fp64") 
plt.legend()
plt.ylim([0,0.6])
plt.title("NumPy MatMul FP64")
plt.ylabel("TFLOPS")
plt.xlabel("Matrix Size")
plt.show()
```

```python
accelerate_fp16 = results["accelerate"]["half"]
conda_fp16 = results["conda"]["half"]
x_axis_fp16 = x_axis[:len(conda_fp16)]
plt.plot(x_axis_fp16, accelerate_fp16, label="accelerate fp16") 
plt.plot(x_axis_fp16, conda_fp16, label="conda fp32") 
plt.legend()
plt.ylim([0,0.001])
plt.title("NumPy MatMul FP16")
plt.ylabel("TFLOPS")
plt.xlabel("Matrix Size")
plt.show()
```
