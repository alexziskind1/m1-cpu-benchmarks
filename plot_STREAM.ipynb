```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
ryzen_x = [1,2,3,4,6,12,20]
ryzen_y = [42335.9, 43593.8, 42724.3, 41935.3, 40705.1, 36847.3, 37569.0]
ryzen_y = np.asarray(ryzen_y)/1e3

m1_x = [1,2,4,8,9,10,18,20]
m1_y = [97039.2, 177503.5, 175829.3, 170873.0, 186633.1, 199212.6, 199477.3, 200677.3]
m1_y = np.asarray(m1_y)/1e3
```

```python
plt.title("Peak memory bandwidth (STREAM)")
plt.ylabel("Memory bandwidth (GB/s)")
plt.xlabel("Number of threads")
plt.plot(ryzen_x, ryzen_y, label="5600X")
plt.plot(m1_x, m1_y, label="M1 Max")
plt.ylim(0, 224)
plt.legend()
plt.show()
```
